```python
from fastai.vision.all import *
path = "gdrive/MyDrive/dope"
```

```python
def is_yummy(x): return x[0].isupper()
dls = ImageDataLoaders.from_name_func(
    path, get_image_files(path), valid_pct=0.2, seed=42,
    label_func=is_yummy, item_tfms=Resize(224))

learn = cnn_learner(dls, resnet34, metrics=error_rate)
learn.fine_tune(3)
```

```python

#hide_output
uploader = widgets.FileUpload()
uploader
```

```python
img = PILImage.create(uploader.data[0])

# rozpoznajemy obrazek we wczytanym modelu
is_yummy_var,_,probs = learn.predict(img)

print(f"Is this yummy?: {is_yummy_var}.")
print(f"Probability it's yummy: {probs[1].item():.6f}")
```
